```
csv_url = 'https://raw.githubusercontent.com/Chokerino/ER-Stress-Analysis-Prediction_Tool/master/gene_expression.csv'
```

```
%matplotlib inline
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv(csv_url)
sns.set(font_scale=0.5)
countplt = sns.countplot(x="stage", data=df, palette='hls')
plt.show()
```

```
max_gene=[]
header=list(df.head(0))

for row in range(0,df.shape[0]):
  temp_max=-10
  max_gene_name=""
  for column in range(1,df.shape[1]-1):

    if temp_max<df.iloc[row,column]:
      temp_max=df.iloc[row,column]
      max_gene_name=header[column]

  max_gene.append(max_gene_name)

max_gene
```

```
df=df.replace(to_replace="Stage I", value=0)
df=df.replace(to_replace="Stage IA", value=1)
df=df.replace(to_replace="Stage IB", value=2)
df=df.replace(to_replace="Stage II", value=3)
df=df.replace(to_replace="Stage IIA", value=4)
df=df.replace(to_replace="Stage IIB", value=5)
df=df.replace(to_replace="Stage III", value=6)
df=df.replace(to_replace="Stage IIIA", value=7)
df=df.replace(to_replace="Stage IIIB", value=8)
df=df.replace(to_replace="Stage IV", value=9)
sns.set(font_scale=1.5)
countplt = sns.countplot(x="stage", data=df, palette='hls')
plt.show()
```

```
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2)
feature_cols=[]
for i in train.columns.values:
    if i != "Patient Name" and i!= "stage":
        feature_cols.append(i)
```

```
from sklearn.svm import LinearSVC
X_train = train.loc[:, feature_cols]
Y_train = train.loc[:, ["stage"]]
model = LinearSVC(C=0.1)
model.fit(X_train, Y_train.values.ravel())
```

```
from sklearn.metrics import accuracy_score
X_test = test.loc[:, feature_cols]
Y_test = test.loc[:, ["stage"]]
Y_pred = model.predict(X_test)
acc = accuracy_score(Y_pred, Y_test.values.ravel())
print(acc)
```
